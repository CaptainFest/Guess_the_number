```python
import sys
MAX = 10**5

def query(TEMP, j, K):
    ans = [0 for i in range(8)]
    rez = TEMP
    for i in range(1,9):
        pr = TEMP+(MAX/8**j * i)
        print(pr, '?')
        sys.stdout.flush()
        if pr > K:
            ans[i-1] = 1
        else:
            ans[i-1] = 0
    for i in range(8):
        if ans[i] > ans[i-1]:
            rez =TEMP+ MAX/8**j*i
    print('+')
    sys.stdout.flush()
    return rez

if __name__ == '__main__':
    TEMP = 0
    K = 100000
    rez = 0
    ap = [0 for i in range(4)]
    for i in range(1,6):
        TEMP = query(TEMP,i,K)
    T = round(TEMP)
    for i in range(4):
        print(T+i,'?')
        sys.stdout.flush()
        if T+i > K:
            ap[i] = 1
        else:
            ap[i] = 0
    print('+')
    if all([a==0 for a in ap]):
        rez = T+3
    else:
        for i in range(4):
            if ap[i] > ap[i-1]:
                rez = T+i-1
    print(str(rez)+'!')
    sys.stdout.flush()
```
